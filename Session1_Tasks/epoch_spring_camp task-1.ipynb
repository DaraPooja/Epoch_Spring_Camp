```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load image as grayscale
image = np.array(Image.open('greypic.jpg'))

# Get image dimensions
h, w = image.shape[:2]

# Split into quadrants
top_left = image[:h//2, :w//2]   # Original
top_right = np.fliplr(top_left)  # Flip horizontally
bottom_left = np.flipud(top_left)  # Flip vertically
bottom_right = np.flipud(top_right)  # Flip vertically (since top_right was flipped horizontally)

# Combine quadrants to form the kaleidoscope effect
top_half = np.hstack((top_left, top_right))
bottom_half = np.hstack((bottom_left, bottom_right))
fimage = np.vstack((top_half, bottom_half))

# Display the result
plt.imshow(fimage, cmap='gray')
plt.axis("off")
plt.show()
```
